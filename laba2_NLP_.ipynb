```python

import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F



import IPython

from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio, load_voice, load_voices

# Это приведет к загрузке всех моделей, используемых Tortoise, из хаба HuggingFace.
tts = TextToSpeech()
```

```python
# Это текст, который будет произнесен.
text = "Витя, братан! Погода сегодня хорошая. Какие у тебя планы на выходные? Пойдем в кино?»"

# Выберите «предустановленный режим» для определения качества. Опции: {"ultra_fast", "fast" (по умолчанию), "standard", "high_quality"}.
preset = "fast"
```

```python
# При желании загрузите свой собственный голос, запустив следующие две ячейки. Я рекомендую
# вы загружаете как минимум 2 аудиоклипа. Это должен быть файл WAV длиной 6–10 секунд.
CUSTOM_VOICE_NAME = "IVAN_1"

import os
from google.colab import files

custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder)
for i, file_data in enumerate(files.upload().values()):
  with open(os.path.join(custom_voice_folder, f'{i}.wav'), 'wb') as f:
    f.write(file_data)
```

```python
# Генерация речи с помощью пользовательского голоса.
voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                          preset=preset)
torchaudio.save(f'generated-{CUSTOM_VOICE_NAME}.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-{CUSTOM_VOICE_NAME}.wav')
```
